# SUPPORT VIRTUAL MACHINE 
USE FOR HAND DIGITS,IMAGE CLASIFICATION AND BIOLOGICAL PROTEIN

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
sf=load_iris()
dir(sf)
```

```python
#here we make irisi data frame
dfm=pd.DataFrame(sf.data,columns=sf.feature_names)
dfm.head()
```

```python
dfm['taget']=sf.target
dfm.head()
```

```python
dfm['flower']=dfm.taget.apply(lambda x :sf.target_names[x] )
dfm.head()
```

```python
dfm[:55]
#dfm[50:100]
#dfm[100:]
```

```python
df1=dfm[:50]
df2=dfm[50:100]
df3=dfm[100:150]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('sepal lengh')
plt.ylabel('sepal width')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='red')
```

```python
plt.xlabel('petal lengh')
plt.ylabel('petal width')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='red')
```

```python
x_h=dfm.drop(['taget','flower'],axis='columns')
x_h.head()
```

```python
y_h=dfm.taget
y_h.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x_h,y_h,test_size=0.2)
```

```python
from sklearn.svm import SVC
mods=SVC()
```

```python
mods.fit(x_h,y_h)
```

```python
mods.predict(X_test)
```

```python
mods.predict([[4.8,3.0,1.5,0.3]])
```

```python
mods.score(X_test,y_test)
```

# here we use some method to improve accuracy Regulaization,Gamma,Kernal

```python
#REGULIZATION
reg=SVC(C=2)
reg.fit(X_train,y_train)
reg.score(X_train,y_train)
```

```python
#now GAMMA PARAMENTER
reg=SVC(gamma=10)
reg.fit(X_train,y_train)
reg.score(X_train,y_train)
```

```python
reg=SVC(kernel='linear')
reg.fit(X_train,y_train)
reg.score(X_train,y_train)
```
